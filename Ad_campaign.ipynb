# **CTR ANALYSIS**

```python
print(df.columns)
```

```python
df_clean = df[df['impressions'] > 0].copy()
```

```python
df_clean['CTR'] = df_clean['clicks'] / df_clean['impressions']
print(df_clean)
```

### **CTR STATISTIC ANALYSIS**

```python
ctr_clean_summary = df_clean['CTR'].describe()
print(ctr_clean_summary)
```

```python
# Top and bottom 5 CTR records (valid data only)
top_ctr_clean = df_clean.sort_values(by='CTR', ascending=False).head(5)
bottom_ctr_clean = df_clean.sort_values(by='CTR', ascending=True).head(5)
```

```python
print(top_ctr_clean[['age','gender', 'impressions','clicks', 'CTR']])
```

```python
print(bottom_ctr_clean[['age', 'gender', 'impressions', 'clicks', 'CTR']])
```

```python
ctr_clean_summary, top_ctr_clean[['ad_id', 'age', 'gender', 'impressions', 'clicks', 'CTR']], bottom_ctr_clean[['ad_id', 'age', 'gender', 'impressions', 'clicks', 'CTR']]
```

```python
# Plot CTR distribution
plt.figure(figsize=(10, 6))
sns.histplot(df_clean['CTR'], bins=50, kde=True, color='steelblue')
plt.title("CTR (Click Through Rate) Distribution", fontsize=14)
plt.xlabel("CTR")
plt.ylabel("Frequency")
plt.grid(True)
plt.tight_layout()
plt.show()
```

**AVERAGE  CTR BY AGE GROUPS**



```python
age_ctr = df_clean.groupby('age')['CTR'].mean().reset_index()

plt.figure(figsize=(8, 5))
sns.barplot(x='age', y='CTR', data=age_ctr, palette='Blues_d')
plt.title('Average CTR by Age Group')
plt.xlabel('Age Group')
plt.ylabel('Average CTR')
plt.tight_layout()
plt.show()
```

**CTR BY GENDER**



```python
# Calculate average CTR by gender
gender_ctr = df_clean.groupby('gender')['CTR'].mean().reset_index()

# Plot
plt.figure(figsize=(6, 4))
sns.barplot(x='gender', y='CTR', data=gender_ctr, palette='Set3')
plt.title('Average CTR by Gender')
plt.xlabel('Gender')
plt.ylabel('Average CTR')
plt.tight_layout()
plt.show()
```

**CTR BY OVERTIME**

```python
# Make sure 'reporting_start' is a datetime column
df_clean['reporting_start'] = pd.to_datetime(df_clean['reporting_start'])

# Group by date and calculate average CTR
time_ctr = df_clean.groupby('reporting_start')['CTR'].mean().reset_index()

# Plot CTR over time
plt.figure(figsize=(10, 5))
sns.lineplot(data=time_ctr, x='reporting_start', y='CTR', marker='o', color='teal')
plt.title('CTR Over Time')
plt.xlabel('Date')
plt.ylabel('Average CTR')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

**CTR VS SPENT**

```python
# Scatter plot: CTR vs. Spent
plt.figure(figsize=(8, 5))
sns.scatterplot(data=df_clean, x='spent', y='CTR', hue='gender', alpha=0.7)
plt.title('CTR vs. Spent')
plt.xlabel('Amount Spent ($)')
plt.ylabel('Click Through Rate (CTR)')
plt.grid(True)
plt.tight_layout()
plt.show()
```

# **CPC (Cost Per Click) Analysis**

```python
# CPC (Cost per Click)
df['CPC'] = df['spent'] / df['clicks']
print(df['CPC'])
```

```python
df_cpc = df_clean[df_clean['clicks'] > 0].copy()
print(df_cpc)
```

```python
df_cpc['CPC'] = df_cpc['spent'] / df_cpc['clicks']
```

```python
cpc_summary = df_cpc['CPC'].describe()
print(cpc_summary)
```

```python
top_5_cpc = df_cpc.sort_values(by='CPC', ascending=True).head(5)
print(top_5_cpc[['ad_id', 'gender', 'age', 'CPC']])
```

```python
bottom_5_cpc = df_cpc.sort_values(by='CPC', ascending=False).head(5)
print(bottom_5_cpc[['ad_id', 'gender', 'age', 'CPC']])
```

**CPC Distribution Plot**

```python
plt.figure(figsize=(10, 6))
sns.histplot(df_cpc['CPC'], bins=40, kde=True, color='slateblue')
plt.title('Distribution of CPC (Cost Per Click)')
plt.xlabel('CPC ($)')
plt.ylabel('Frequency')
plt.grid(True)
plt.tight_layout()
plt.show()
```

**CPC by Age & Gender**

```python
# Average CPC by age and gender
age_gender_cpc = df_cpc.groupby(['age', 'gender'])['CPC'].mean().reset_index()

plt.figure(figsize=(10, 6))
sns.barplot(data=age_gender_cpc, x='age', y='CPC', hue='gender', palette= ['#CBAACB', '#A3E4D7'])

plt.title('Average CPC by Age and Gender')
plt.xlabel('Age Group')
plt.ylabel('Average CPC')
plt.tight_layout()
plt.show()
```

**CPC OVERTIME**

```python
# CPC over time
df_cpc['reporting_start'] = pd.to_datetime(df_cpc['reporting_start'])
time_cpc = df_cpc.groupby('reporting_start')['CPC'].mean().reset_index()

plt.figure(figsize=(10, 5))
sns.lineplot(data=time_cpc, x='reporting_start', y='CPC', marker='o', color='darkgreen')
plt.title('CPC Over Time')
plt.xlabel('Date')
plt.ylabel('Average CPC')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

**CPC VS SPENT**

```python
plt.figure(figsize=(8, 5))
sns.scatterplot(data=df_cpc, x='spent', y='CPC', hue='gender', alpha=0.7)
plt.title('CPC vs. Amount Spent')
plt.xlabel('Amount Spent ($)')
plt.ylabel('CPC ($)')
plt.grid(True)
plt.tight_layout()
plt.show()
```
